### Agregar un elemento al final del arreglo

```python
var estaciones = ['Invierno', 'Otoo', 'Primavera'];
estaciones.push('Verano');
console.log(estaciones);
//Consola: ['Invierno', 'Otoo', 'Primavera', 'Verano'] El mtodo push() aade un elemento al final del arreglo.
```

### Remover el ltimo elemento 

```python
var estaciones
estaciones = ['Invierno', 'Otoo', 'Primavera', 'Verano'];
estaciones.pop();
console.log(estaciones);
//Consola: ['Invierno', 'Otoo', 'Primavera'] 
//	El mtodo pop() elimina el ltimo elemento del arreglo.
// Tambien puede agregar el elemento removido a una variable
var miEstacion = estaciones.pop();
console.log(miEstacion);
//Consola: 'Verano' 
// Esto almacena el elemento removido en la variable miEstacion.
```

### Remover el primer elemento 


```python
var estaciones = ['Invierno', 'Otoo', 'Primavera', 'Verano'];
estaciones.shift();
console.log(estaciones);
//Consola: ['Otoo', 'Primavera', 'Verano'] 
// El mtodo shift() elimina el primer elemento del arreglo.
```

### Agregar un elementos al inicio


```python
var estaciones = ['Invierno', 'Otoo', 'Primavera'];
estaciones.unshift('Verano');
console.log(estaciones);
//Consola: ['Verano', 'Invierno', 'Otoo', 'Primavera'] 
// El mtodo unshift() aade un elemento al inicio del arreglo.
```
