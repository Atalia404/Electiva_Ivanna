### Primer Hola Mundo con Node.js
* Crear un archivo con la .js al final.


```python
var http = require('http');

http.createServer(function (req, res) {
    res.writeHead(200, {'Content-Type': 'text/html'});
    res.end('Hello World!');
}).listen(8080);
```

* require('http'): Imagina que require es como un "importador" que trae funciones especiales a tu programa. En este caso, require('http') trae todas las herramientas necesarias para trabajar con el protocolo HTTP, que es como los navegadores web se comunican con los servidores. Al usarlo, obtienes un objeto que te permite crear servidores web y manejar solicitudes.

* Creaci칩n del Servidor: Luego, se crea una aplicaci칩n de servidor. Piensa en esto como levantar un "recepcionista" que estar치 atento a las peticiones que lleguen. Cada vez que alguien (un cliente, como un navegador web) hace una solicitud a tu aplicaci칩n, el recepcionista activa una funci칩n espec칤fica para manejarla.

* Manejo de Solicitudes (writeHead y end): Cuando el servidor recibe una solicitud, la funci칩n que se activa se encarga de preparar una respuesta. En este ejemplo, la respuesta es un simple "Hola mundo".

*      writeHead: Antes de enviar el mensaje, el servidor necesita enviar informaci칩n adicional sobre la respuesta, como qu칠 tipo de contenido es (por ejemplo, texto plano) y si todo sali칩 bien (c칩digo 200 OK). Esto es lo que hace writeHead: env칤a los "encabezados" de la respuesta al cliente.

*      end: Despu칠s de enviar el mensaje y los encabezados, la funci칩n end se encarga de finalizar la comunicaci칩n y cerrar la conexi칩n con el cliente.

* server.listen(8080): Esta funci칩n es como "abrir la puerta" de tu servidor en un n칰mero espec칤fico, que se llama "puerto". En este caso, el puerto es el 8080. Cuando tu servidor "escucha" en el puerto 8080, significa que est치 listo para recibir solicitudes de cualquier cliente que intente conectarse a esa direcci칩n y puerto. Puedes elegir cualquier puerto que no est칠 siendo usado.

> En resumen, el c칩digo te permite construir un peque침o servidor web que, cuando alguien lo visita, le responde con un "Hola mundo".

    ### Crear un modulo y Importar y Exportar 
    

    > Un modulo en la practica es un archivo de Javascript, se crea un archivo: saludo.js y otro modulo llamado app.js.                             
    > El primer modulo ser치 creado con la definici칩n de la funcion y el segundo modulo es para llamar a la funcion.

```python
// saludo.js
funtion saludar(nombre) {
    return `Hola ${nombre}`;
}
//exportacion
console.log(module.exports); // solo es de prueba
module.exports.saludar = saludar; 

//exportar otra funcion
function saludarHolaMundo() {
    return "Hola Mundo";
}
module.exports.saludarHolaMundo = saludarHolaMundo;


// Importar varios elementos en uno:
module.exports = {
    saludar: saludar,
    saludarHolaMundo: saludarHolaMundo
};
```

```python
// app.js

//importar el modulo
const saludo = require("./saludo.js");
console.log(saludo.saludar("Ivanna"));
//_________(objeto.propiedad asociada a la funcion).

// importar otra funcion
console.log(saludo.saludarHolaMundo());
```

### Sintaxis de desestructuraci칩n y require()


```python
// app.js
// Importar solo una funci칩n espec칤fica
const { saludarHolaMundo } = require("./saludo.js"); 
// or asi para agregar m치s funciones
const { saludar, saludarHolaMundo } = require("./saludo.js"); 
console.log(saludo.saludarHolaMundo());
```
