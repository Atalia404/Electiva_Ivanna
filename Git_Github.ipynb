# GIT
Controlador de versiones de proyectos local que se tenga en la computadora.
GitHub
Es donde se van a sincronizar los cambios en el internet 
Instalacin: 
Descargar Git desde su pgina oficial y configurar en el cmd con:
- git config --global user.name _________(nombre de usuario de github)
- git config --global user.email _____ (emanil de github)

Para saber si la configuracin esta bien se utiliza el comando: 
- git config --global --list (aparecera el user) 

Status de git
- git status

### Inicializar un repositorio 
- git init 									
		
Revisar que todos los archivos en la barra izquierda del vsc estn en verde, esto indica que puedes trackearlos al repositorio.

### Para agregar un archivo en especfico:
- git add + nombre del archivo a agregar.
- git add . (Para agregar todos los archivos se usa el ".")

### Commits
Luego de eso puedes hacer un commit, que no es mas que darle el nombre a los cambios que se esta haciendo a esta versin 
- git commit -m "lo que se quiera poner version base" 

### Comando para saber cuantos commits se han hecho en el proyecto, se usa:
- git log

Una vez que se agregan, se trackean los archivos, estos ya no se vern en verde  en la barra del explorador de archivos en el vsc, una vez que se hace una modificacin en algn archivo , se colorea su nombre en azul aguamarina  y se realiza de nuevo el add y el commit.
- git add.
- git commit -m "cambie el fondo a azul" 
- git log para saber cuantas versiones van

### Para volver a la ltima versin que se le hizo commit, por algn error que no se puede resolver, se usa: 
- git stash 

Los buenos commits tienen una ciencia de como puede ser llamados o descritos: *Convencional Commit*			

- [tipo](alcance):[asunto][descripcin]

Ejemplo: Se debe hacer un cambio de un componente - Buttom a color Naranja. El commit se describira:

> (Refactor)(Buttom.jsx):(cambiar de color del boton)(cambiar color del fondo del componente a naranja de azl por medio de la clase css de Tailwinds)

- Tipo: fix (agregas algo), test, refactor, revert, build... feat (archivo nuevo)

- Alcance: Nombre del archivo que fue modificado
- Asunto: El nombre del commit que debe ser escrito en Presente Perfecto
- Descripcin:Por qu y el cmo? del cambio realizado (Resolver, cambiar, actualizar...) 

### Hay archivos que no queremos que se suban al repositorio y con ello se usa: 
- git ignore

# GITHUB

- Ir a GitHub.
- Crear un nuevo repositorio.
- Colocar descripcin.
- Elegir si es un proyecto pblico o privado.
- Elegir si se le agrega un README
- Te dan dos opciones subir un repositorio nuevo desde cero O subir un repositorio que ya se tiene hecho. 
- Al tener un repositorio se puede subir con los comandos dados en la pgina del repositorio:

	1) La primera linea de comando es para setear o sincronizar el repositorio local con el repositorio remoto
	2) La segunda linea de comando: 
	- git branch -m main 
	3) El ltimo comando se realiza el push de la rama main al repositorio externo 
	- git push -u origin main

Si es la primera vez hay que logearse a nuestra cuenta de GitHub, y luego al terminar la sincronizacin, se observarn todos los archivos en la pgina del GitHub

### Modificacin de un archivo git a github nuevo:
- git add . 
- git commit -m "...."
- git push 
Con eso se reflejan los cambios en GitHub.
