### Generador de colsulta

Ver todos los campos y todos los registros de una tabla: 
- SELECT * FROM tabla	*(nombre de la tabla a consultar)*

Ver uno o varios campos y todos los registros de una tabla: 
- SELECT campo1, campo2 FROM tabla;

Agregarle un alias a los campos de nuestra consulta
- SELECT campo1 as aliasCampo1, campo2 as aliasCampo2 from tabla;

Filtrar la consulta por los valores de uno o varios campos
- SELECT * FROM tabla WHERE campo1='valor1' and campo2>=9;

Ordenar los registros de nuestra tabla de forma ascendente y descendente
- SELECT * from tabla order by nota desc;

Ordenar los valores minimos y maximos de un campo de una tabla:
- SELECT min(campo) as aliasMin, max(campo) as aliasMax from tabla;
Ejemplo: 
- SELECT 'alumnos'.'nombre','alumnos'.'nota'
	FROM 'alumnos'
	ORDER BY nota asc;

Limita la cantidad de registros a mostrar:
- SELECT * FROM tabla limit 3,2;

Buscar datos que contenga una letra en especÃ­fico: 
- SELECT 'alumnos'.'nombre', 'alumno'.'fecha_ingreso' 
	FROM 'alumnos' 
	WHERE 'alumnos'.'nombre' *LIKE %J%* AND 'alumnos'.'notas' >= '0';
